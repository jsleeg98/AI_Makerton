### Ïù¥ Îïå Îç∞Ïù¥ÌÑ∞Í∞Ä ÏûàÎäî Ìè¥ÎçîÎ°ú ÏúÑÏπòÎ°ú pathÎ•º Î∞îÍøîÏ£ºÏñ¥ÏïºÌïúÎã§.
### path Ìè¥Îçî ÏïàÏóêÎäî Ïù¥ÎØ∏ÏßÄÍ∞Ä Îì§Ïñ¥ÏûàÎäî images Ìè¥ÎçîÏôÄ txtÎ°ú ÎùºÎ≤®ÎßÅÏù¥ Îì§Ïñ¥ÏûàÎäî labels Ìè¥ÎçîÏôÄ data.yaml ÌååÏùºÏù¥ ÏûàÏñ¥ÏïºÌïúÎã§.

```python
path = '/home/twopiece/datasets_rw/final-dataset/'

from glob import glob

img_list = glob(path + 'images/*.png')

print(len(img_list))

from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size = 0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))

with open(path + 'train.txt', 'w') as f:
    f.write('\n'.join(train_img_list) + '\n')
    
with open(path + 'val.txt','w') as f:
    f.write('\n'.join(val_img_list)+ '\n')

import yaml

with open(path + 'data.yaml', 'r') as f:
    data = yaml.load(f)
    
print(data)

data['train'] = path + 'train.txt'
data['val'] = path + 'val.txt'

with open(path + 'data.yaml', 'w') as f:
    yaml.dump(data,f)

print(data)
```
